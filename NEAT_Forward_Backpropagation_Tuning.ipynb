```python
#Étape Optionelle: suppression des données aberrantes.

import pandas as pd
import numpy as np

def drop_outliers_IQR(df):
    """
    Drops outliers from a DataFrame using the IQR method.

    Returns:
        pandas.DataFrame: The DataFrame with outliers removed.
    """
    # Select only numeric columns for outlier detection
    numeric_cols = df.select_dtypes(include=np.number).columns

    # Calculate quantiles and IQR only for numeric columns
    q1 = df[numeric_cols].quantile(0.25)
    q3 = df[numeric_cols].quantile(0.75)
    IQR = q3 - q1

    # Identify outliers based on numeric columns
    not_outliers = df[~((df[numeric_cols] < (q1 - 1.5 * IQR)) | (df[numeric_cols] > (q3 + 1.5 * IQR))).any(axis=1)]

    # Drop rows with missing values and reset index
    outliers_dropped = not_outliers.dropna().reset_index(drop=True)

    return outliers_dropped

df1 = df
df_outliers_dropped = drop_outliers_IQR(df1)
df = df_outliers_dropped
df_outliers_dropped
```

```python
#Version NEAT sans généralisation et optimisation.

import neat
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Splittimg data into training and testing.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Define fitness function
def eval_genome(genome, config):
    net = neat.nn.FeedForwardNetwork.create(genome, config)
    correct = 0
    for xi, target in zip(X_train, y_train):
        output = net.activate(xi)
        if np.argmax(output) == target:
            correct += 1
    return correct / len(y_train)  # Accuracy as fitness

best_fitness_log = []

def eval_genomes(genomes, config):
    global best_fitness_log
    for genome_id, genome in genomes:
        net = neat.nn.FeedForwardNetwork.create(genome, config)
        fitness = eval_genome(genome, config)
        genome.fitness = fitness
    best_fitness_log.append(max([g.fitness for _, g in genomes]))

# Configure NEAT
config_path = "./neat_cfg_vis.txt"
config = neat.Config(neat.DefaultGenome, neat.DefaultReproduction,
                     neat.DefaultSpeciesSet, neat.DefaultStagnation, config_path)

# Run NEAT
pop = neat.Population(config)
stats = neat.StatisticsReporter()
pop.add_reporter(stats)
pop.add_reporter(neat.StdOutReporter(True))

winner = pop.run(eval_genomes, 100)

# Get the best model (winner)
net = neat.nn.FeedForwardNetwork.create(winner, config)

# Predict on test data
y_pred = []
for x in X_test:
    output = net.activate(x)
    predicted_class = np.argmax(output)  # Get class with highest probability
    y_pred.append(predicted_class)

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("NEAT Confusion Matrix")
plt.show()
```

```python

def draw_net(genome, config, filename="network.png"):
    """ Visualize a neural network genome structure. """
    net = neat.nn.FeedForwardNetwork.create(genome, config)

    G = nx.DiGraph()

    m_net = net

    # Add nodes and assign colors
    # input nodes are blue, output nodes are red, and hidden nodes are gray
    node_colors = {}
    for i in net.input_nodes:
        G.add_node(i)
        node_colors[i] = 'yellow'  # Input layer
    for i in net.output_nodes:
        G.add_node(i)
        node_colors[i] = 'green'  # Output layer
    for conn in genome.connections.values():
        if conn.enabled:
            G.add_edge(conn.key[0], conn.key[1], weight=conn.weight)
            if conn.key[0] not in node_colors:  # Check if node is hidden
                G.add_node(conn.key[0])
                node_colors[conn.key[0]] = 'gray'
            if conn.key[1] not in node_colors:  # Check if node is hidden
                G.add_node(conn.key[1])
                node_colors[conn.key[1]] = 'gray'

    # Draw graph using node_colors for node color mapping
    colors = [node_colors[n] for n in G.nodes]
    pos = nx.spring_layout(G)
    nx.draw(G, pos, node_color=colors, with_labels=True, edge_color='gray', alpha=0.8)
    plt.savefig(filename)
    plt.show()

draw_net(winner, config)
```

```python
#Version NEAT généraliser et optimiser - 100 generations

import neat
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import json
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from torch.utils.data import DataLoader, TensorDataset
import json
import os

#Reset complexity log file
if not os.path.exists("complexity_log.json"):
    with open("complexity_log.json", "w") as f:
        f.write("")

X_fit, X_val, y_fit, y_val = None, None, None, None

# --- Neural Network Model ---
class EvolvedNN(nn.Module):
    def __init__(self, neat_net, genome):
        super().__init__()
        self.layers = nn.ModuleList()
        self.node_map = {}
        layer_index = 0
        neuron_index = 0

        self.dropout = nn.Dropout(p=0.2)  # Add dropout layer

        input_size = len(neat_net.input_nodes)
        self.layers.append(nn.Linear(input_size, len(neat_net.input_nodes)))

        for node_id in neat_net.input_nodes:
            self.node_map[node_id] = (layer_index, neuron_index)
            neuron_index += 1

        hidden_nodes = [n for n in genome.nodes if n not in neat_net.input_nodes and n not in neat_net.output_nodes]
        num_hidden = len(hidden_nodes)

        if num_hidden > 0:
            layer_index += 1
            neuron_index = 0
            self.layers.append(nn.Linear(len(neat_net.input_nodes), num_hidden))
            for node_id in hidden_nodes:
                self.node_map[node_id] = (layer_index, neuron_index)
                neuron_index += 1

        layer_index += 1
        neuron_index = 0
        input_size = num_hidden if num_hidden > 0 else len(neat_net.input_nodes)
        self.layers.append(nn.Linear(input_size, len(neat_net.output_nodes)))
        for node_id in neat_net.output_nodes:
            self.node_map[node_id] = (layer_index, neuron_index)
            neuron_index += 1

        self.load_neat_weights(genome)

    def forward(self, x):
        for i, layer in enumerate(self.layers):
            x = layer(x)
            if i < len(self.layers) - 1 and i > 0:
                x = torch.relu(x)
        return x

    def load_neat_weights(self, genome):
        for conn in genome.connections.values():
            if conn.enabled:
                in_node, out_node = conn.key
                weight = conn.weight
                in_layer, in_neuron = self.node_map.get(in_node, (None, None))
                out_layer, out_neuron = self.node_map.get(out_node, (None, None))
                if in_layer is not None and out_layer is not None and out_layer < len(self.layers):
                    try:
                        self.layers[out_layer].weight.data[out_neuron, in_neuron] = weight
                        self.layers[out_layer].bias.data[out_neuron] = genome.nodes[out_node].bias
                    except IndexError:
                        pass

# --- Training Helper Functions ---

def genome_to_model(genome, config):
    neat_net = neat.nn.FeedForwardNetwork.create(genome, config)
    return EvolvedNN(neat_net, genome)


def fine_tune_model(model, X, y, epochs=3, batch_size=32, lr=0.001):
    model.train()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    criterion = nn.CrossEntropyLoss()
    min_samples = min(X.shape[0], y.shape[0])
    X = X[:min_samples]
    y = y[:min_samples]
    dataset = TensorDataset(torch.tensor(X, dtype=torch.float32), torch.tensor(y.values, dtype=torch.long))
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    for _ in range(epochs):
        for inputs, targets in loader:
            optimizer.zero_grad()
            outputs = torch.log_softmax(model(inputs), dim=1)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()

# --- Fitness Function with Enhancements ---
def smooth_fitness(previous_fitness, new_fitness, alpha=0.85):
    return alpha * new_fitness + (1 - alpha) * previous_fitness

# Delete fitness_log
if os.path.exists("fitness_log.json"):
    os.remove("fitness_log.json")

def fitness_function(genomes, config):
    models = []
    complexity_data = []
    for genome_id, genome in genomes:
        model = genome_to_model(genome, config)
        models.append((genome_id, genome, model))

    models.sort(key=lambda tup: tup[1].fitness if tup[1].fitness is not None else 0.0, reverse=True)
    top_k = int(len(models) * 0.2)

    raw_fitness_log = []
    smoothed_fitness_log = []

    global X_fit, X_val, y_fit, y_val

    # Split training set into train/validation for generalization
    X_fit, X_val, y_fit, y_val = train_test_split(X_train, y_train, test_size=0.3)
    y_val = y_val.reset_index(drop=True)  # Reset index to avoid potential mismatches

    for i, (genome_id, genome, model) in enumerate(models):
        if i < top_k:
            fine_tune_model(model, X_fit, y_val, epochs=3)

        model.eval()
        with torch.no_grad():
            inputs = torch.tensor(X_test, dtype=torch.float32)
            targets = torch.tensor(y_test.values, dtype=torch.long)
            outputs = model(inputs)
            predictions = outputs.argmax(1).numpy()
            accuracy = accuracy_score(y_test, predictions)

        # Apply complexity penalty
        penalty = 0.001# or 0.005
        raw_fitness = max(0.0, accuracy - penalty)

        # Apply fitness smoothing
        if genome.fitness is not None:
            smoothed = smooth_fitness(genome.fitness, raw_fitness)
        else:
            smoothed = raw_fitness

        genome.fitness = smoothed

        # Record complexity
        complexity_data.append({
            "genome_id": genome_id,
            "connections": len(genome.connections),
            "nodes": len(genome.nodes),
            "raw_fitness": raw_fitness,
            "smoothed_fitness": smoothed
        })

    # Save complexity log per generation
    with open("complexity_log.json", "a") as f:
        f.write(json.dumps(complexity_data) + "\n")
        raw_fitness_log.append(raw_fitness)
        smoothed_fitness_log.append(smoothed)

    with open("fitness_log.json", "a") as f:
        json.dump({"raw": raw_fitness_log, "smoothed": smoothed_fitness_log}, f)
        f.write("\n")

# --- Run NEAT ---
config = neat.Config(
    neat.DefaultGenome,
    neat.DefaultReproduction,
    neat.DefaultSpeciesSet,
    neat.DefaultStagnation,
    './neat_cfg_vis.txt'
)

population = neat.Population(config)

stats = neat.StatisticsReporter()
population.add_reporter(stats)
population.add_reporter(neat.StdOutReporter(True))

winner = population.run(fitness_function, 100)

# --- Final Backprop on Winner Genome ---
neat_net = neat.nn.FeedForwardNetwork.create(winner, config)
model = EvolvedNN(neat_net, winner)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

train_loss_log = []
train_acc_log = []

for epoch in range(100):
    model.train()
    total_loss = 0
    correct = 0
    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = torch.log_softmax(model(inputs), dim=1)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        correct += (outputs.argmax(1) == targets).sum().item()

    epoch_loss = total_loss / len(train_loader)
    epoch_acc = correct / len(train_loader.dataset)
    train_loss_log.append(epoch_loss)
    train_acc_log.append(epoch_acc)

# Save training logs
with open("backprop_log.json", "w") as f:
    json.dump({"loss": train_loss_log, "accuracy": train_acc_log}, f)

# Final evaluation
y_pred_tensor = model(X_test_tensor)
y_pred = torch.argmax(y_pred_tensor, axis=1).numpy()
test_accuracy = accuracy_score(y_test, y_pred)
print(f"Final Accuracy on Test Set: {test_accuracy:.4f}")
```

```python

import pickle
import neat
import matplotlib.pyplot as plt
from collections import defaultdict

# Load config
config_path = "neat_cfg_vis.txt"
config = neat.Config(
    neat.DefaultGenome,
    neat.DefaultReproduction,
    neat.DefaultSpeciesSet,
    neat.DefaultStagnation,
    config_path
)

# Load evolved winner genome
#evolved_neat_model.pkl
#winner_genome.pkl
with open("evolved_neat_model.pkl", "rb") as f:
    winner = pickle.load(f)

# Compute input sensitivity (sum of absolute weights)
input_sensitivity = defaultdict(float)
for (input_node, output_node), conn in winner.connections.items():
    if conn.enabled and input_node < 0:  # NEAT uses negative IDs for input nodes
        input_sensitivity[input_node] += abs(conn.weight)

# Input label mapping
input_labels = {
    -1: "avg_F1",
    -2: "jitter_s",
    -3: "shimmer",
    -4: "mean_hnr"
}

# Prepare data
labels = [input_labels[i] for i in sorted(input_sensitivity.keys())]
sensitivities = [input_sensitivity[i] for i in sorted(input_sensitivity.keys())]

# Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(labels, sensitivities, color='lightcoral')
plt.title("Input Sensitivity (Sum of Absolute Connection Weights)")
plt.ylabel("Total Influence")
plt.xlabel("Input Features")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# Upload the complexity log from NEAT run

# Parse complexity_log.json
generation_data = []
with open("complexity_log.json", "r") as f:
    for line in f:
        try:
            gen = json.loads(line)
            generation_data.append(gen)
        except json.JSONDecodeError:
            continue

# Aggregate average complexity and fitness per generation
avg_connections = []
avg_nodes = []
avg_raw_fitness = []
avg_smoothed_fitness = []

for gen in generation_data:
    if not gen: continue
    avg_connections.append(sum(g['connections'] for g in gen) / len(gen))
    avg_nodes.append(sum(g['nodes'] for g in gen) / len(gen))
    avg_raw_fitness.append(sum(g['raw_fitness'] for g in gen) / len(gen))
    avg_smoothed_fitness.append(sum(g['smoothed_fitness'] for g in gen) / len(gen))

generations = list(range(1, len(avg_connections) + 1))

# Plot: Accuracy vs. Connections
plt.figure(figsize=(10, 5))
plt.plot(generations, avg_raw_fitness, label='Raw Fitness', color='blue')
plt.plot(generations, avg_smoothed_fitness, label='Smoothed Fitness', color='green')
plt.xlabel("Generation")
plt.ylabel("Fitness (Accuracy)")
plt.title("Fitness Over Generations")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot: Connections vs. Fitness
plt.figure(figsize=(10, 5))
plt.plot(avg_connections, avg_raw_fitness, 'o-', label='Raw Fitness')
plt.plot(avg_connections, avg_smoothed_fitness, 'x-', label='Smoothed Fitness')
plt.xlabel("Avg. Number of Connections")
plt.ylabel("Fitness (Accuracy)")
plt.title("Fitness vs. Network Complexity (Connections)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# evolved_model.py or app.py

import neat
import pickle
import torch.nn as nn
import torch

# Load NEAT config
config = neat.Config(
    neat.DefaultGenome,
    neat.DefaultReproduction,
    neat.DefaultSpeciesSet,
    neat.DefaultStagnation,
    "neat_cfg_vis.txt"
)

# Load winner genome
with open("winner_genome.pkl", "rb") as f:
    winner = pickle.load(f)

# Define the evolved PyTorch model
class EvolvedNN(nn.Module):
    def __init__(self, neat_net, genome):
        super().__init__()
        self.layers = nn.ModuleList()
        self.node_map = {}
        layer_index = 0
        neuron_index = 0

        for node_id in neat_net.input_nodes:
            self.node_map[node_id] = (layer_index, neuron_index)
            neuron_index += 1

        hidden_nodes = [n for n in genome.nodes if n not in neat_net.input_nodes and n not in neat_net.output_nodes]
        num_hidden = len(hidden_nodes)

        if num_hidden > 0:
            layer_index += 1
            neuron_index = 0
            self.layers.append(nn.Linear(len(neat_net.input_nodes), num_hidden))
            for node_id in hidden_nodes:
                self.node_map[node_id] = (layer_index, neuron_index)
                neuron_index += 1

        layer_index += 1
        neuron_index = 0
        input_size = num_hidden if num_hidden > 0 else len(neat_net.input_nodes)
        self.layers.append(nn.Linear(input_size, len(neat_net.output_nodes)))
        for node_id in neat_net.output_nodes:
            self.node_map[node_id] = (layer_index, neuron_index)
            neuron_index += 1

        self.load_neat_weights(genome)

    def forward(self, x):
        for i, layer in enumerate(self.layers):
            x = layer(x)
            if i < len(self.layers) - 1 and i > 0:
                x = torch.relu(x)
        return x

    def load_neat_weights(self, genome):
        for conn in genome.connections.values():
            if conn.enabled:
                in_node, out_node = conn.key
                weight = conn.weight
                in_layer, in_neuron = self.node_map.get(in_node, (None, None))
                out_layer, out_neuron = self.node_map.get(out_node, (None, None))
                if in_layer is not None and out_layer is not None and out_layer < len(self.layers):
                    try:
                        self.layers[out_layer].weight.data[out_neuron, in_neuron] = weight
                        self.layers[out_layer].bias.data[out_neuron] = genome.nodes[out_node].bias
                    except IndexError:
                        pass
```
