```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

#Defining fuzzy variables
quality=ctrl.Antecedent(np.arange(0,11,1), 'quality')
service=ctrl.Antecedent(np.arange(0,11,1), 'service')
tip=ctrl.Consequent(np.arange(0,26,1),'tip')

#Defining membership functions
quality['low'] = fuzz.trimf(quality.universe,[0,0,5])
quality['medium'] = fuzz.trimf(quality.universe,[0,5,10])
quality['high'] = fuzz.trimf(quality.universe,[5,10,10])

service['low'] = fuzz.trimf(service.universe,[0,0,5])
service['medium'] = fuzz.trimf(service.universe,[0,5,10])
service['high'] = fuzz.trimf(service.universe,[5,10,10])

tip['low'] = fuzz.trimf(tip.universe,[0,0,13])
tip['medium'] = fuzz.trimf(tip.universe,[0,13,25])
tip['high'] = fuzz.trimf(tip.universe,[13,25,25])

#Define the rules
rule1 = ctrl.Rule(quality['high'] | service['high'], tip['high'])
rule2 = ctrl.Rule(service['medium'], tip['medium'])
rule3 = ctrl.Rule(quality['low'] | service['low'], tip['low'])

#Compile the rules into a working model
tip_ctrl = ctrl.ControlSystem([rule1,rule2,rule3])
tip_simulation = ctrl.ControlSystemSimulation(tip_ctrl)

#Set the input variables values
tip_simulation.input['quality'] = 6.5
tip_simulation.input['service'] = 9.8

#Given the input variables and the fuzzy logic rules applied, the output is computed (tip amount)
tip_simulation.compute()

#Print the output (tip amount)
print(f"The tip is {tip_simulation.output['tip']}")
```
